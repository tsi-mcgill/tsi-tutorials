## Functions

Let's write a function that will be vectorized by numpy

````python
def get_sqrt_real(x):
    if x > 0:
        return np.sqrt(x)
    else:
        return np.nan
````

If we try to run this on a numpy array, it wont work

````python
x_sqrt = get_sqrt_real(x)
````

Instead we can "vectorize" the function using `np.vectorize`

````python
get_sqrt_real_vec = np.vectorize(get_sqrt_real)
````

````python
x_sqrt = get_sqrt_real_vec(x)
````

````python
plt.plot(x, x_sqrt)
plt.xlim(x[0], x[-1])
````
